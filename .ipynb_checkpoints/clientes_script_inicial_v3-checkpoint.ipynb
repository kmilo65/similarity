```python
#import mysql.connector
from sqlalchemy import create_engine
import pymysql
import numpy as np
import pandas as pd
from numpy.random import default_rng
```

```python
#cnx = sqlalchemy.create_engine('mysql+pymysql://root:chilote64@localhost/analitica2')
cnx = create_engine('mysql+pymysql://root:chilote64@localhost/analitica2')
```

```python
qry="select	clientcode,hiercode,sum(nc) as nc from 	rp_client_preference_hst group by clientcode,hiercode"
df_input=pd.DataFrame(cnx.execute(qry).fetchall())
df_input.columns=['clientcode','hiercode','nc']
```

```python
df_input.columns=['clientcode','hiercode','nc']
```

```python
# entire dataframe
df_input
```

```python
df_input.shape
```

```python
lst_client=list(set(df_input['clientcode'].tolist()))
len(lst_client)
```

```python
list_client_array=np.array(lst_client).reshape((-1,1))
```

```python
list_client_array
```

```python
print(len(list_client_array))
```

```python
# Number of clients to choose randomly
n=20000
```

```python
rng = default_rng()
random_array = rng.choice(len(lst_client), size=n, replace=False)
```

```python
random_array
```

```python
random_array.shape
```

```python
# convering the index array to a list
mask=random_array.tolist()
mask
```

```python
len(mask)
```

```python
# randonly choose clients 
list_client=list_client_array[mask].reshape((1,-1)).tolist()[0]
len(list_client)
```

```python
# convert the list to a dataframe
client_df=pd.DataFrame(list_client)
client_df.columns=['clientcode']
client_df=client_df.sort_values(by='clientcode', ascending=True)
client_df.shape
```

```python
client_df
```

```python
# describing the dataframe
client_df.describe()
```

```python
# index for clients
client_df.index = pd.RangeIndex(len(client_df))
client_df.index
idx_client=pd.Series(client_df.index, index=client_df['clientcode'])
idx_client.shape
```

```python
idx_client
```
